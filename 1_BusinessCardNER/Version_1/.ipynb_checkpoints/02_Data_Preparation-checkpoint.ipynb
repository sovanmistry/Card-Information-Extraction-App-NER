```python
import numpy as np
import pandas as pd
import cv2
import pytesseract

import os
from glob import glob
from tqdm import tqdm

import warnings
warnings.filterwarnings('ignore')
```

```python
imgPaths = glob('./Selected/*.jpeg')
```

```python
imgPaths
```

## Extract FileName

```python
imgPath = imgPaths[0]
print(imgPath)
```

```python
_, filename = os.path.split(imgPath)
print(filename)
```

## Extract Info for 1 Business Card

```python
image = cv2.imread(imgPath)
data = pytesseract.image_to_data(image)
dataList = list(map(lambda x : x.split('\t'), data.split('\n')))
df = pd.DataFrame(dataList[1:], columns=dataList[0])

# Dropping Null Values
df.dropna(inplace = True)
df['conf'] = df['conf'].astype(int)

# take only those data where confidence Score > 30
useFulData = df.query('conf > 30')

# DataFrame
businessCard = pd.DataFrame()
businessCard['text'] = useFulData['text']
businessCard['id'] = filename
```

```python
businessCard.head()
```

## Extract info for all Business Card

```python
allBusinessCard = pd.DataFrame(columns = ['id', 'text'])


for imgPath in tqdm(imgPaths, desc = 'BusinessCard'):

    # imgPath = imgPaths[0]
    # print(imgPath)

    _, filename = os.path.split(imgPath)
    # print(filename)

    image = cv2.imread(imgPath)
    data = pytesseract.image_to_data(image)
    dataList = list(map(lambda x : x.split('\t'), data.split('\n')))
    df = pd.DataFrame(dataList[1:], columns=dataList[0])

    # Dropping Null Values
    df.dropna(inplace = True)
    df['conf'] = df['conf'].astype(int)

    # take only those data where confidence Score > 30
    useFulData = df.query('conf > 30')

    # DataFrame
    businessCard = pd.DataFrame()
    businessCard['text'] = useFulData['text']
    businessCard['id'] = filename
    
    # Concatenate 
    allBusinessCard = pd.concat((allBusinessCard, businessCard))
```

```python
allBusinessCard 
```

## Save Data into CSV

```python
allBusinessCard.to_csv('businessCardDemo.csv', index = False)
```
