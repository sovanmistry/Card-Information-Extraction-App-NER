## Image to Data

```python
data = pytesseract.image_to_data(img_cv)
```

```python
data
```

```python
print(data)
```

```python
dataList = list(map(lambda x : x.split('\t'), data.split('\n')))
df = pd.DataFrame(dataList[1:], columns=dataList[0])
```

```python
df.head()
```

### **Conclusion**:
    
**level 1** : From where the Page  started

**level 2** : From where the block started

**level 3** : From where the para  started

**level 4** : From where the line  started

**level 5** : From where the Word  started

```python
df.info()
```

### **Conclusion :**
    
1. We need to delete the Null Values
2. all NON-Text data should be converted into Int

### **Drop Missing Values**

```python
df.dropna(inplace = True)
```

### **Change the Data Type from Object to Integer**

```python
col_int = ['level', 'page_num', 'block_num', 'par_num', 'line_num', 'word_num', 'left', 'top', 'width', 'height', 'conf']
df[col_int] = df[col_int].astype(int)
```

```python
df.info()
```

## **Draw Bounding Box Around Each Word**

```python
image = img_cv.copy()
level = 'word'

for l,x,y,w,h,c,txt in df[['level','left', 'top', 'width', 'height', 'conf', 'text']].values:
    
    if level == 'page':
        if l == 1:
            cv2.rectangle(img = image, pt1 = (x,y), pt2 = (x+w, y+h), color = (0,0,0), thickness=2)
        else : 
            continue
            
    
    elif level == 'block':
        if l == 2:
            cv2.rectangle(img = image, pt1 = (x,y), pt2 = (x+w, y+h), color = (255,0,0), thickness=2)
        else : 
            continue
            
            
    elif level == 'para':
        if l == 3:
            cv2.rectangle(img = image, pt1 = (x,y), pt2 = (x+w, y+h), color = (0,0,0), thickness=2)
        else : 
            continue
            
    elif level == 'line':
        if l == 4:
            cv2.rectangle(img = image, pt1 = (x,y), pt2 = (x+w, y+h), color = (0,0,0), thickness=2)
        else : 
            continue
            
            
    elif level == 'word':
        if l == 5:
            cv2.rectangle(img = image, pt1 = (x,y), pt2 = (x+w, y+h), color = (0,255,0), thickness=2)
            cv2.putText(image, txt,(x,y), cv2.FONT_HERSHEY_PLAIN, 1, (255,0,0), 2)
        else : 
            continue
            
cv2.imshow("bounding box", image)
cv2.waitKey()
cv2.destroyAllWindows()
```
